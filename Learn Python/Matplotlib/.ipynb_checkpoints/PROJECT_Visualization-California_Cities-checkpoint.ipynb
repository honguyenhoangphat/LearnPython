# PROJECT: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
cities = pd.read_csv('california_cities.csv')
```

```python
cities.head(-5)
#latitude /n/: Vĩ độ
#longtitude /n/: Kinh độ
```

```python
# Extract Latd = Vĩ độ and Longd = Kinh đọ

lat, long = cities['latd'], cities['longd']

#Plot using Pyplot API
plt.scatter(lat, long)

plt.axis('equal');
plt.xlabel('longtitude');
plt.ylabel('latitude');
```

```python
# Extract Latd = Vĩ độ and Longd = Kinh đọ

long, lat = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

plt.figure(figsize=(10,6))
plt.style.use('seaborn-v0_8')
#Plot using Pyplot API
plt.scatter(lat, long,
           c=np.log10(population),
           cmap='viridis',
           s=area, linewidths=0, alpha=0.5)
# log10 làm số dân giảm xuống log10
plt.axis('equal');
plt.xlabel('longtitude');
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$(population)');
plt.clim(3,7);

plt.title('California Cities: Population and Area Distribution')

#Create a legend for Cities SIZE
area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s=area, c='k',
               label=str(area) + 'km$^2$')
plt.legend(labelspacing=1, title='City Area');
```
