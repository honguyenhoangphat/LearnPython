### 2 cách: Pylot API vs Object-Oriented (OO) API
- Pyplot API -> Quickly
- OO API -> Advanced

Pyploy API

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), "blue", linestyle="dashed", label= "sin(x)")
plt.plot(x, np.cos(x), "red", label = "cos(x)")

plt.title("Sin(x) and Cos(x)")
plt.xlabel("Biến x")
plt.ylabel("F(x)")

#plt.xlim([0,6]) # Giới hạn x
#plt.ylim([-0.75, 0.75])
#plt.axis(x(min), x(max), y(min), y(max))
#plt.axis("tight")
#plt.axis("equal")

#Để in label lên, thì dùng legend
plt.legend();
```

### Object-Oriented API

```python
#1. Prepare data
x= [1,2,3,4]
y= [11,22,33,30]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,y);
ax.set(title='A simple plot',
       xlabel = 'x-axis',
       ylabel = 'y-axis');
```

### Line

```python
x = np.linspace(0,10,100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

### Scatter

```python
#Pyplot API
plt.scatter(x, np.exp(x));
```

```python
## OO API

rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
sizes = 1000*rng.rand(100)
colors = rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s=sizes, c=colors, cmap='viridis', alpha = 0.4);

fig.colorbar(img1);
```

### BAR
* Vertical
* Horizontal

```python
# Prepare Data from a Python dictionary
soft_drink_prices = {"Coca": 10,
                     "Peppi": 12,
                     "Sprai": 7}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title="Winmart Softdrink Price", ylabel="Price");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()));
# LẬT NGƯỢC CÁI ĐỒ THỊ LẠI
```

## HISTOGRAM

```python
np.random.seed(42)
student_height = np.random.normal(160, 20, 250)
```

```python
#pyplot API
plt.hist(student_height,bins = 15);
```

## SUBPLOTS

```python
# Option 1:
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5));
ax1.plot(x, x/2);
ax1.plot(x, x*2);
ax2.scatter(np.random.random(500), np.random.random(500));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins=30);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5));
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);
ax[0,1].scatter(np.random.random(500), np.random.random(500));
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1,1].hist(student_height, bins=30);
```
