# Variance PHƯƠNG SAI
    Là phép đo mức chênh lệch giữa các số liệu trong một tập dữ liệu trong thống kê.
    Nghĩa là giá trị kỳ vọng của bình phương của độ lệch của X so với giá trị trung bình của nó.
    Trong một tập dữ liệu, bạn sẽ trừ từng giá trị khỏi giá trị trung bình riêng lẻ, sau đó bình phương giá trị, (μ – X) ². Sau đó, bạn sẽ cộng tất cả các độ lệch bình phương lại với nhau và chia chúng cho tổng số giá trị để đạt được mức trung bình. Con số này là phương sai.
Var (X) = E (x – μ) ² / N

Trong đó:

x là giá trị trong tập dữ liệu

µ là giá trị trung bình của tập dữ liệu

n là số lượng giá trị trong tập dữ liệu    


Giá trị trung bình của các con chó


Lấy giá trị của con chó đó TRỪ cho giá trị trung bình. Sau đó bình phương độ lệch rồi chia cho số số lượng con chó



Độ lệch chuẩn là căn bậc hai của phương sai. Kẽ hai đường hai bên giá trị trung bình

```python
# Mô tả ví dụ các con chó vào code
dog_height = [600,470,170,430,300]
dog_height = np.array(dog_height)
np.std(dog_height)
```

```python
np.var(dog_height)
```

```python
np.sqrt(np.var(dog_height)) # DO LECH CHUAN
```
